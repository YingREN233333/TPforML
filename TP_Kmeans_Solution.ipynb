```python
#Saisir votre code ici
from skimage import io
import matplotlib.pyplot as plt
image1 = io.imread(r'C:\Users\zhaddad\Downloads\lung.png',as_gray=True)/255.0
image2 = io.imread(r'C:\Users\zhaddad\Downloads\beach4.jpg')/255.0
print(image1.shape)
print(image2.shape)
plt.figure()
plt.imshow(image1)
plt.figure()
plt.imshow(image2)
```

<div class="alert alert-block alert-info" style='color:MidnightBlue; background-color: AliceBlue ; border-radius: 30px; border-color: black; '>
<hr><h2>2) Application de l’algorithme Kmeans:</h2><hr> 
a) Redimensionnez les images en deux dimensions (largeur*hauteur, 1) pour l'image médicale et (largeur*hauteur, 3) pour l'autre.<br>
b) Importez la fonction KMeans à partir de sklearn.cluster<br>
c) Appliquez un clustering Kmeans sur les données de l’image en utilisant 3 clusters.<br>
d) Calculez les valeurs des images segmentées via le clustering Kmeans crée.<br>
</div>

```python

from sklearn.cluster import KMeans
h1, w1 = image1.shape
h2, w2, c = image2.shape
image1_2d = image1.reshape(h1*w1,1)
image2_2d = image2.reshape(h2*w2, 3)

kmeans1 = KMeans(n_clusters=3)
kmeans1.fit(image1_2d)
cluster_centers1 = kmeans1.cluster_centers_
cluster_labels1 = kmeans1.labels_
print(cluster_centers1)
print(cluster_labels1)

kmeans2 = KMeans(n_clusters=3, random_state=0).fit(image2_2d)
image2_result = kmeans2.cluster_centers_[kmeans2.labels_]
```

<div class="alert alert-block alert-info" style='color:MidnightBlue; background-color: AliceBlue ; border-radius: 30px; border-color: black; '>

```python

import matplotlib.pyplot as plt

newimage1 = cluster_centers1[cluster_labels1].reshape(h1, w1)*255.0
newimage1 = newimage1.astype('uint8')
plt.imshow(image1)
plt.show() 
io.imshow(newimage1)
io.show()
io.imsave('lung_kmeans.gif', newimage1)


image2_result = image2_result.reshape(h2, w2, 3)
plt.imshow(image2)
plt.show() 
plt.imshow(image2_result)
plt.show() 
```
