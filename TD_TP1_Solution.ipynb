```python
#1)chargement des données
import pandas as pd
path="C:/EPF/2022_2023/MIN/IAII/séance2/TP2/donnees_TP2.csv"
df = pd.read_csv(path)
#error_bad_lines=False
#selection de la première colonne de notre dataset (la taille de la population)
X = df.iloc[0:len(df),0]
X=X.values.reshape(len(X),1)
#selection de deuxième colonnes de notre dataset (le profit effectué)
Y = df.iloc[0:len(df),1] 
Y=Y.values.reshape(len(Y),1)

#2)visualisation des données
import matplotlib.pyplot as plt
axes = plt.axes()
axes.grid() # dessiner une grille pour une meilleur lisibilité du graphe
plt.scatter(X,Y) # X et Y sont les variables qu'on a extraite dans le paragraphe précédent


#3)Apliquer l'algo
import sklearn.linear_model
# Selection d'un modèle linéaire
model = sklearn.linear_model.LinearRegression()
# Entrainement du modèle
model.fit(X, Y)


#4) effectuer des prédictions
X_new = [[21.72]]  
print("prédiction = ",model.predict(X_new))
# retourne : 22.14674235

#5)
import numpy as np
X_b = np.c_[np.ones(np.shape(X)), X]  # ajouter x0 = 1 à chaque observation

theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)
print("theta_best = ", theta_best)

X_new_b = np.c_[np.ones(np.shape(X_new)), X_new]  # ajouter x0 = 1 à chaque observation
y_predict = X_new_b.dot(theta_best)
print("y_predict =",y_predict)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#génération de données
np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)
plt.scatter(x,y, s=50 )
plt.show()

#appliquer une regression linéaire
from sklearn.linear_model import LinearRegression

# transformer les données pour ajouter un nouvea axe
x = x[:, np.newaxis]
y = y[:, np.newaxis]

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

plt.scatter(x, y, s=10)
plt.plot(x, y_pred, color='r')
plt.show()
#calcul de l'erreur quadratique moyenne et du score R2
from sklearn.metrics import mean_squared_error, r2_score
rmse = np.sqrt(mean_squared_error(y,y_pred))
r2 = r2_score(y,y_pred)
print("rmse = ",rmse)
print("r2 = ",r2)

#appliquer une regression polynomiale
import operator
from sklearn.preprocessing import PolynomialFeatures

#appliquer une transformation polynomialede degré 2 aux données x pour avoir x_poly
#polynomial_features= PolynomialFeatures(degree=2)
#polynomial_features= PolynomialFeatures(degree=3)
for nbr in [2,3,20]:
    polynomial_features= PolynomialFeatures(degree=nbr)
    x_poly = polynomial_features.fit_transform(x)
    #print("x = ",x)
    #print("x_poly = ",x_poly)

    model = LinearRegression()
    model.fit(x_poly, y)
    y_poly_pred = model.predict(x_poly)

    rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
    r2 = r2_score(y,y_poly_pred)
    print("rmse = ",rmse)
    print("r2 = ",r2)

    plt.scatter(x,y)
    # trier les valeurs de x avant le tracé
    sort_axis = operator.itemgetter(0)
    sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
    x, y_poly_pred = zip(*sorted_zip)
    plt.plot(x, y_poly_pred, color='m')
    plt.show()

#Avec un degree=20 on est face à un phénomène d'overfitting (surajustement)
```

Pour résoudre ce problème (le phénomène observé), nous devons augmenter la complexité du
modèle. Pour générer une équation d'ordre supérieur, nous transformons le modèle linéaire<br><br> 
Y = Ɵ<sub>0</sub>+ Ɵ<sub>1</sub> X<br>
en :<br> 
Y = Ɵ<sub>0</sub>+ Ɵ<sub>1</sub> X+ Ɵ<sub>2</sub> X<sup>2</sup><br><br> 
Il faut savoir que ce modèle est toujours considéré comme un modèle linéaire car les
coefficients/poids associés aux caractéristiques sont toujours linéaires. En effet, X² peut être vu
aussi comme une caractéristique. Toutefois, notons bien que la courbe que nous ajustons est bien
de nature quadratique.<br>
Pour convertir les caractéristiques originales X en caractéristiques d’ordre supérieur, nous pouvons
utiliser la classe PolynomialFeatures fournie par scikit-learn. Nous utiliserons alors les deux instructions
suivantes :<br>
<CODE>

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

def create_polynomial_regression_model(degree,X_train,Y_train,X_test,Y_test):
  
  poly_features = PolynomialFeatures(degree=degree)
  
  # transformer les entrées X en X polynomiales.
  X_train_poly = poly_features.fit_transform(X_train)
  
  # créer le modèle  Linear Regression et faire l'apprentissage sur les données
  poly_model = LinearRegression()
  poly_model.fit(X_train_poly, Y_train)
  
  # prediction sur le data-set d'apprenitssage
  y_train_predicted = poly_model.predict(X_train_poly)
  
  # prediction sur le data-set du test
  y_test_predict = poly_model.predict(poly_features.fit_transform(X_test))
  
  # evaluation du modèle sur le dataset d'apprentissage 
  rmse_train = np.sqrt(mean_squared_error(Y_train, y_train_predicted))
  r2_train = r2_score(Y_train, y_train_predicted)
  
  # evaluation du modèle sur le dataset de test
  rmse_test = np.sqrt(mean_squared_error(Y_test, y_test_predict))
  r2_test = r2_score(Y_test, y_test_predict)
  
  print("Mesures de performance du modèle sur les données d'apprentissage")
  print("-------------------------------------------")
  print("RMSE sur la base d'apprentissage = {}".format(rmse_train))
  print("R2 sur la base d'apprentissage = {}".format(r2_train))
  
  print("\n")
  
  print("Mesures de performance du modèle sur les données de test")
  print("-------------------------------------------")
  print("RMSE sur la base de test = {}".format(rmse_test))
  print("R2 sur la base de test = {}".format(r2_test))
```
